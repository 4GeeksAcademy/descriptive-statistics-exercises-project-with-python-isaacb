```python
# grafica tus resultados

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Cargar dataset de IMDb
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)

df['title_length'] = df['title'].apply(len)
media = np.mean(df['title_length'])
mediana = np.median(df['title_length'])
moda = mode(df['title_length'])
fig, ax = plt.subplots(figsize=(12, 7))

# Histograma media y mediana
ax.hist(df['title_length'], bins=25, color='#69b3a2', edgecolor='black', alpha=0.7)
ax.axvline(media, color='#e74c3c', linestyle='--', linewidth=3, label=f'Media: {media:.2f}')
ax.axvline(mediana, color='#f1c40f', linestyle='--', linewidth=3, label=f'Mediana: {mediana}')
ax.set_title('Distribución de la longitud de títulos (IMDb)', fontsize=18, fontweight='bold')
ax.set_xlabel('Longitud del título (número de caracteres)', fontsize=14)
ax.set_ylabel('Frecuencia', fontsize=14)
legend = ax.legend(frameon=True, shadow=True, fontsize=12)
frame = legend.get_frame()
frame.set_facecolor('white')
frame.set_edgecolor('black')
plt.tight_layout()
plt.show()


## grafico de rango desviacion standart

valores = [rango, ds, varianza]
etiquetas = ['Rango', 'Desviación estándar', 'Varianza']
plt.figure(figsize=(8,5))
bars = plt.bar(etiquetas, valores, color=['#ff6f61', '#6b5b95', '#88b04b'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{yval:.2f}', ha='center', fontsize=12)

plt.title('Comparación de Rango, Desviación estándar y Varianza', fontsize=16, fontweight='bold')
plt.ylabel('Valor')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

#grafico asimetría y curtosis de la columna 'title_length'
valores = [asimetria, kts]
etiquetas = ['Asimetría', 'Curtosis']

plt.figure(figsize=(6,5))
bars = plt.bar(etiquetas, valores, color=['#e67e22', '#9b59b6'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'{yval:.3f}', ha='center', fontsize=12)

plt.title('Asimetría y Curtosis de la longitud de títulos', fontsize=16, fontweight='bold')
plt.ylabel('Valor')
plt.ylim(min(0, min(valores)) - 0.5, max(valores) + 0.5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Ejercicio 2 – Cálculo manual de desviación estándar

En este ejercicio vas a calcular manualmente la desviación estándar, seleccionando cinco títulos reales del dataset de películas que cargaste antes.


```python
# Crea una lista con 5 valores reales de df['title_length'], por ejemplo: [10, 13, 14, 18, 22]
valores_reales = df['title_length'].sample(n=5, random_state=42).tolist()
```

```python
# Calcula la media de los 5 valores
valores_reales = df['title_length'].sample(n=5, random_state=42).tolist()
media = np.mean(valores_reales)
media
```

```python
# Calcula las diferencias al cuadrado con respecto a la media
diferencias_cuadrado = [(x - media) ** 2 for x in valores_reales]
diferencias_cuadrado
```

```python
# Calcula la varianza (promedio de las diferencias al cuadrado) y su raíz cuadrada para obtener la desviación
varianza = sum(diferencias_cuadrado) / len(valores_reales)
desviacion_estandar = varianza ** 0.5
```
